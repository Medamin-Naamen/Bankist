```python
# Liste pour stocker les données, X_train, X_test, y_train, y_test
X_train = pd.DataFrame()
X_test  = pd.DataFrame()
y_train = pd.Series()
y_test  = pd.Series()

# Boucle pour traiter chaque ensemble de données de d1 à d90
for i in range(1, 91):
    # Charger les données
    data = globals()[f"df{i}"]

    # Séparer les caractéristiques (X) et les étiquettes (Y)
    X = data.iloc[:, 0:-1]
    Y = data.iloc[:, -1]

    # Diviser les données en ensembles d'entraînement et de test
    X_tr, X_ts, y_tr, y_ts = train_test_split(X, Y, test_size=0.25, random_state=42)

    # Concaténer les données d'entraînement et de test dans les DataFrames
    X_train = pd.concat([X_train, X_tr], ignore_index=True)
    X_test = pd.concat([X_test, X_ts], ignore_index=True)
    y_train = pd.concat([y_train, y_tr], ignore_index=True)
    y_test = pd.concat([y_test, y_ts], ignore_index=True)
```
